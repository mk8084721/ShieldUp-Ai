```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Configuration
NUM_SESSIONS_PER_CLASS = 500
TIMESTEPS = 30
```

```python

# ============================================================================
# overlap and noise
# ============================================================================

def add_temporal_pattern(base_values, pattern_type='stable'):
    """Add realistic temporal evolution with MORE variance"""
    result = np.array(base_values, dtype=float)

    if pattern_type == 'stable':
        # More noise than before
        noise = np.random.normal(0, 0.25, len(result))  # Was 0.1, now 0.25
        return result * (1 + noise)

    elif pattern_type == 'ramp_up':
        ramp = np.linspace(0.3, 1.5, len(result))  # Wider range
        noise = np.random.normal(0, 0.15, len(result))
        return result * ramp * (1 + noise)

    elif pattern_type == 'burst':
        result = result.copy()
        num_bursts = np.random.randint(2, 6)  # More bursts
        for _ in range(num_bursts):
            burst_start = np.random.randint(0, len(result) - 5)
            burst_len = np.random.randint(2, 8)
            result[burst_start:burst_start + burst_len] *= np.random.uniform(1.8, 4.0)
        return result

    elif pattern_type == 'decay':
        decay = np.linspace(1.5, 0.4, len(result))
        noise = np.random.normal(0, 0.2, len(result))
        return result * decay * (1 + noise)

    elif pattern_type == 'stealth':
        stealth = np.concatenate([
            np.linspace(0.2, 0.4, len(result)//2),
            np.linspace(0.5, 2.0, len(result)//2)  # Bigger spike
        ])
        return result * stealth

    elif pattern_type == 'chaotic':
        # New: Highly variable behavior
        chaos = np.random.uniform(0.3, 2.0, len(result))
        return result * chaos

    return result

def add_feature_noise(value, noise_level=0.35):  # Increased from 0.15
    """Add MORE realistic noise"""
    # Random activation
    if value == 0 and np.random.random() > 0.90:  # Was 0.95, now 0.90
        return np.random.uniform(0, 5)

    # Much more multiplicative noise
    noise = np.random.uniform(1 - noise_level, 1 + noise_level)
    return max(0, value * noise)

def generate_correlated_features(base_intensity, feature_config):
    """Generate features with MORE independence and variance"""
    features = {}

    # Camera features - more independent
    cam_factor = np.random.uniform(0.4, 1.8)  # Was 0.7-1.3, now wider
    features['cam_count'] = feature_config['cam_count'] * base_intensity * cam_factor
    features['cam_duration'] = feature_config['cam_duration'] * base_intensity * cam_factor
    features['cam_bg_ratio'] = feature_config['cam_bg_ratio'] * base_intensity

    # Location features - highly independent
    loc_factor = np.random.uniform(0.3, 2.0)  # Was 0.7-1.3
    features['loc_freq'] = feature_config['loc_freq'] * base_intensity * loc_factor
    features['loc_bg_ratio'] = feature_config['loc_bg_ratio'] * base_intensity

    # Microphone - very independent
    mic_factor = np.random.uniform(0.2, 2.5)  # Was 0.5-1.5
    features['mic_duration'] = feature_config['mic_duration'] * base_intensity * mic_factor
    features['mic_bg_flag'] = 1 if (base_intensity > 0.4 and np.random.random() > 0.25) else 0

    # Data features - can vary wildly
    data_factor = np.random.uniform(0.3, 2.2)  # Was 0.6-1.4
    features['data_upload'] = feature_config['data_upload'] * base_intensity * data_factor
    features['data_download'] = feature_config['data_download'] * base_intensity
    features['bg_data'] = feature_config['bg_data'] * base_intensity

    features['reserved'] = 0

    return features
```

```python

# ============================================================================
# CLASS OVERLAP
# ============================================================================

def generate_safe_session():
    """Safe with MUCH more edge cases - 30% instead of 10%"""

    is_edge_case = np.random.random() < 0.30  # Was 0.10!

    if is_edge_case:
        edge_type = np.random.choice([
            'video_filter', 'ar_game', 'educational',
            'photo_app', 'voice_game', 'streaming'  # New edge cases
        ])

        if edge_type == 'video_filter':
            base_config = {
                'cam_count': np.random.uniform(2, 5),  # Higher
                'cam_duration': np.random.uniform(15, 40),  # MUCH higher - overlaps suspicious
                'cam_bg_ratio': np.random.uniform(0, 0.25),  # Some background
                'loc_freq': np.random.uniform(0, 2),
                'loc_bg_ratio': np.random.uniform(0, 0.2),
                'mic_duration': np.random.uniform(0, 15),
                'data_upload': np.random.uniform(10, 80),  # Higher
                'data_download': np.random.uniform(30, 200),
                'bg_data': np.random.uniform(0, 20)
            }
            pattern = np.random.choice(['stable', 'burst', 'chaotic'])

        elif edge_type == 'ar_game':
            base_config = {
                'cam_count': np.random.uniform(3, 6),  # Very high
                'cam_duration': np.random.uniform(20, 50),  # Overlaps suspicious heavily
                'cam_bg_ratio': np.random.uniform(0.1, 0.35),
                'loc_freq': np.random.uniform(2, 6),
                'loc_bg_ratio': np.random.uniform(0.1, 0.3),
                'mic_duration': np.random.uniform(5, 25),
                'data_upload': np.random.uniform(20, 120),
                'data_download': np.random.uniform(80, 300),
                'bg_data': np.random.uniform(5, 40)
            }
            pattern = 'burst'

        elif edge_type == 'educational':
            base_config = {
                'cam_count': np.random.uniform(1, 3),
                'cam_duration': np.random.uniform(10, 30),
                'cam_bg_ratio': np.random.uniform(0.05, 0.25),
                'loc_freq': np.random.uniform(2, 5),
                'loc_bg_ratio': np.random.uniform(0.1, 0.3),
                'mic_duration': np.random.uniform(10, 35),
                'data_upload': np.random.uniform(40, 150),
                'data_download': np.random.uniform(80, 300),
                'bg_data': np.random.uniform(10, 50)
            }
            pattern = 'stable'

        elif edge_type == 'photo_app':
            base_config = {
                'cam_count': np.random.uniform(3, 7),  # Very high camera
                'cam_duration': np.random.uniform(10, 35),
                'cam_bg_ratio': np.random.uniform(0, 0.15),
                'loc_freq': np.random.uniform(1, 4),  # Location tagging
                'loc_bg_ratio': np.random.uniform(0, 0.2),
                'mic_duration': np.random.uniform(0, 5),
                'data_upload': np.random.uniform(50, 200),  # Uploading photos
                'data_download': np.random.uniform(30, 150),
                'bg_data': np.random.uniform(5, 30)
            }
            pattern = 'burst'

        elif edge_type == 'voice_game':
            base_config = {
                'cam_count': np.random.uniform(0, 1),
                'cam_duration': np.random.uniform(0, 8),
                'cam_bg_ratio': np.random.uniform(0, 0.1),
                'loc_freq': np.random.uniform(0, 2),
                'loc_bg_ratio': np.random.uniform(0, 0.15),
                'mic_duration': np.random.uniform(30, 80),  # High mic
                'data_upload': np.random.uniform(40, 180),
                'data_download': np.random.uniform(60, 250),
                'bg_data': np.random.uniform(10, 50)
            }
            pattern = 'stable'

        else:  # streaming
            base_config = {
                'cam_count': np.random.uniform(1, 3),
                'cam_duration': np.random.uniform(5, 20),
                'cam_bg_ratio': np.random.uniform(0.15, 0.35),  # Some background
                'loc_freq': np.random.uniform(0, 2),
                'loc_bg_ratio': np.random.uniform(0, 0.2),
                'mic_duration': np.random.uniform(10, 40),
                'data_upload': np.random.uniform(30, 150),
                'data_download': np.random.uniform(100, 400),  # High download
                'bg_data': np.random.uniform(20, 80),
            }
            pattern = 'stable'

    else:
        # Normal safe app - but with MORE variance
        base_config = {
            'cam_count': np.random.uniform(0, 1.5),  # Was 0-0.5
            'cam_duration': np.random.uniform(0, 12),  # Was 0-5
            'cam_bg_ratio': np.random.uniform(0, 0.2),  # Was 0-0.1
            'loc_freq': np.random.uniform(0, 2.5),  # Was 0-1
            'loc_bg_ratio': np.random.uniform(0, 0.2),  # Was 0-0.12
            'mic_duration': np.random.uniform(0, 10),  # Was 0-3
            'data_upload': np.random.uniform(0, 100),  # Was 0-40
            'data_download': np.random.uniform(10, 200),  # Was 10-100
            'bg_data': np.random.uniform(0, 30)  # Was 0-10
        }
        pattern = np.random.choice(['stable', 'burst', 'decay', 'chaotic'],
                                   p=[0.5, 0.2, 0.15, 0.15])

    # Generate with MORE noise
    session = []
    for t in range(TIMESTEPS):
        intensity = 1.0 + np.random.uniform(-0.4, 0.4)  # Was ±0.2, now ±0.4
        features = generate_correlated_features(intensity, base_config)

        for key in features:
            if key != 'reserved' and key != 'mic_bg_flag':
                features[key] = add_feature_noise(features[key], noise_level=0.35)  # Was 0.2

        session.append(features)

    # Apply temporal pattern
    for key in base_config.keys():
        values = [s[key] for s in session]
        modified = add_temporal_pattern(values, pattern)
        for i, s in enumerate(session):
            s[key] = max(0, modified[i])

    return session, 0

def generate_suspicious_session():
    """Suspicious - NOW MUCH MORE VARIABLE AND OVERLAPPING"""

    behavior_profile = np.random.choice([
        'almost_safe',              # NEW: Really looks safe
        'mostly_safe_with_spikes',
        'moderate_throughout',
        'improving',
        'deteriorating',
        'almost_dangerous',         # NEW: Really looks dangerous
        'highly_variable'           # NEW: Chaotic
    ], p=[0.15, 0.20, 0.25, 0.10, 0.10, 0.10, 0.10])

    if behavior_profile == 'almost_safe':
        # This will be VERY hard to distinguish from safe
        base_config = {
            'cam_count': np.random.uniform(0.5, 3),
            'cam_duration': np.random.uniform(5, 25),  # Overlaps safe heavily
            'cam_bg_ratio': np.random.uniform(0.1, 0.35),
            'loc_freq': np.random.uniform(1, 4),
            'loc_bg_ratio': np.random.uniform(0.1, 0.35),
            'mic_duration': np.random.uniform(5, 30),
            'data_upload': np.random.uniform(30, 180),
            'data_download': np.random.uniform(50, 250),
            'bg_data': np.random.uniform(10, 60)
        }
        pattern = np.random.choice(['stable', 'burst'])

    elif behavior_profile == 'mostly_safe_with_spikes':
        base_config = {
            'cam_count': np.random.uniform(0.8, 3),
            'cam_duration': np.random.uniform(8, 30),
            'cam_bg_ratio': np.random.uniform(0.15, 0.45),
            'loc_freq': np.random.uniform(1.5, 5),
            'loc_bg_ratio': np.random.uniform(0.2, 0.5),
            'mic_duration': np.random.uniform(10, 40),
            'data_upload': np.random.uniform(50, 280),
            'data_download': np.random.uniform(100, 400),
            'bg_data': np.random.uniform(20, 90)
        }
        pattern = 'burst'

    elif behavior_profile == 'moderate_throughout':
        base_config = {
            'cam_count': np.random.uniform(1.5, 4),
            'cam_duration': np.random.uniform(15, 50),  # Wide overlap
            'cam_bg_ratio': np.random.uniform(0.25, 0.6),
            'loc_freq': np.random.uniform(3, 8),
            'loc_bg_ratio': np.random.uniform(0.3, 0.65),
            'mic_duration': np.random.uniform(20, 60),
            'data_upload': np.random.uniform(100, 500),
            'data_download': np.random.uniform(150, 550),
            'bg_data': np.random.uniform(40, 150)
        }
        pattern = 'stable'

    elif behavior_profile == 'improving':
        base_config = {
            'cam_count': np.random.uniform(2, 5),
            'cam_duration': np.random.uniform(20, 55),
            'cam_bg_ratio': np.random.uniform(0.35, 0.65),
            'loc_freq': np.random.uniform(4, 9),
            'loc_bg_ratio': np.random.uniform(0.4, 0.7),
            'mic_duration': np.random.uniform(25, 70),
            'data_upload': np.random.uniform(150, 550),
            'data_download': np.random.uniform(200, 650),
            'bg_data': np.random.uniform(60, 180)
        }
        pattern = 'decay'

    elif behavior_profile == 'deteriorating':
        base_config = {
            'cam_count': np.random.uniform(1, 4),
            'cam_duration': np.random.uniform(10, 45),
            'cam_bg_ratio': np.random.uniform(0.2, 0.55),
            'loc_freq': np.random.uniform(2, 7),
            'loc_bg_ratio': np.random.uniform(0.25, 0.6),
            'mic_duration': np.random.uniform(15, 55),
            'data_upload': np.random.uniform(80, 450),
            'data_download': np.random.uniform(120, 500),
            'bg_data': np.random.uniform(30, 140)
        }
        pattern = 'ramp_up'

    elif behavior_profile == 'almost_dangerous':
        # This will be VERY hard to distinguish from dangerous
        base_config = {
            'cam_count': np.random.uniform(2.5, 5),
            'cam_duration': np.random.uniform(35, 80),  # Overlaps dangerous
            'cam_bg_ratio': np.random.uniform(0.5, 0.8),
            'loc_freq': np.random.uniform(5, 10),
            'loc_bg_ratio': np.random.uniform(0.55, 0.85),
            'mic_duration': np.random.uniform(40, 100),
            'data_upload': np.random.uniform(250, 800),
            'data_download': np.random.uniform(250, 750),
            'bg_data': np.random.uniform(100, 350)
        }
        pattern = np.random.choice(['stable', 'ramp_up'])

    else:  # highly_variable
        base_config = {
            'cam_count': np.random.uniform(1, 4.5),
            'cam_duration': np.random.uniform(10, 60),
            'cam_bg_ratio': np.random.uniform(0.2, 0.7),
            'loc_freq': np.random.uniform(2, 8),
            'loc_bg_ratio': np.random.uniform(0.25, 0.7),
            'mic_duration': np.random.uniform(15, 70),
            'data_upload': np.random.uniform(80, 600),
            'data_download': np.random.uniform(120, 600),
            'bg_data': np.random.uniform(40, 200)
        }
        pattern = 'chaotic'

    # Generate with  noise
    session = []
    for t in range(TIMESTEPS):
        intensity = 1.0 + np.random.uniform(-0.5, 0.5)  # Was ±0.3, now ±0.5!
        features = generate_correlated_features(intensity, base_config)

        for key in features:
            if key != 'reserved' and key != 'mic_bg_flag':
                features[key] = add_feature_noise(features[key], noise_level=0.40)  # Was 0.25

        session.append(features)

    # Apply temporal pattern
    for key in base_config.keys():
        values = [s[key] for s in session]
        modified = add_temporal_pattern(values, pattern)
        for i, s in enumerate(session):
            s[key] = max(0, modified[i])

    return session, 1

def generate_dangerous_session():
    """Dangerous with MORE stealth and variance - 30% stealth instead of 15%"""

    is_stealth = np.random.random() < 0.30  # Was 0.15!

    if is_stealth:
        # Stealth has MORE variance now
        stealth_type = np.random.choice(['gradual', 'sudden', 'intermittent'])

        if stealth_type == 'gradual':
            base_config = {
                'cam_count': np.random.uniform(1.5, 4),  # Starts lower
                'cam_duration': np.random.uniform(15, 70),
                'cam_bg_ratio': np.random.uniform(0.4, 0.8),
                'loc_freq': np.random.uniform(3, 9),
                'loc_bg_ratio': np.random.uniform(0.5, 0.85),
                'mic_duration': np.random.uniform(25, 90),
                'data_upload': np.random.uniform(150, 700),
                'data_download': np.random.uniform(120, 550),
                'bg_data': np.random.uniform(60, 350)
            }
            pattern = 'stealth'

        elif stealth_type == 'sudden':
            base_config = {
                'cam_count': np.random.uniform(2, 5),
                'cam_duration': np.random.uniform(20, 90),
                'cam_bg_ratio': np.random.uniform(0.45, 0.85),
                'loc_freq': np.random.uniform(4, 10),
                'loc_bg_ratio': np.random.uniform(0.5, 0.9),
                'mic_duration': np.random.uniform(30, 110),
                'data_upload': np.random.uniform(180, 850),
                'data_download': np.random.uniform(140, 650),
                'bg_data': np.random.uniform(70, 400)
            }
            pattern = 'ramp_up'

        else:  # intermittent
            base_config = {
                'cam_count': np.random.uniform(2, 5),
                'cam_duration': np.random.uniform(25, 85),
                'cam_bg_ratio': np.random.uniform(0.5, 0.85),
                'loc_freq': np.random.uniform(4, 10),
                'loc_bg_ratio': np.random.uniform(0.55, 0.9),
                'mic_duration': np.random.uniform(35, 100),
                'data_upload': np.random.uniform(200, 800),
                'data_download': np.random.uniform(150, 600),
                'bg_data': np.random.uniform(80, 380)
            }
            pattern = 'burst'

    else:
        # Obvious dangerous with  variance
        aggression_level = np.random.uniform(0.6, 1.4)  # Was 0.8-1.2, more spread now

        base_config = {
            'cam_count': np.random.uniform(3, 8) * aggression_level,
            'cam_duration': np.random.uniform(40, 180) * aggression_level,
            'cam_bg_ratio': np.random.uniform(0.55, 0.95),
            'loc_freq': np.random.uniform(6, 15) * aggression_level,
            'loc_bg_ratio': np.random.uniform(0.6, 0.95),
            'mic_duration': np.random.uniform(50, 200) * aggression_level,
            'data_upload': np.random.uniform(350, 1500) * aggression_level,
            'data_download': np.random.uniform(200, 1100) * aggression_level,
            'bg_data': np.random.uniform(120, 700) * aggression_level
        }
        pattern = np.random.choice(['stable', 'ramp_up', 'burst', 'chaotic'],
                                   p=[0.4, 0.3, 0.2, 0.1])

    # Generate with noise
    session = []
    for t in range(TIMESTEPS):
        intensity = 1.0 + np.random.uniform(-0.35, 0.35)  # Was ±0.25
        features = generate_correlated_features(intensity, base_config)

        for key in features:
            if key != 'reserved' and key != 'mic_bg_flag':
                features[key] = add_feature_noise(features[key], noise_level=0.30)  # Was 0.2

        session.append(features)

    # Apply temporal pattern
    for key in base_config.keys():
        values = [s[key] for s in session]
        modified = add_temporal_pattern(values, pattern)
        for i, s in enumerate(session):
            s[key] = max(0, modified[i])

    return session, 2
```

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (LSTM, Dense, Dropout, BatchNormalization,
                                      Bidirectional, GaussianNoise)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l1_l2

# Ensure reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
CSV_PATH = "permission_usage_dataset_v2.csv"

FEATURES = [
    "cam_count", "cam_duration", "cam_bg_ratio",
    "loc_freq", "loc_bg_ratio",
    "mic_duration", "mic_bg_flag",
    "data_upload", "data_download", "bg_data",
    "reserved"
]

TIMESTEPS = 30
NUM_FEATURES = len(FEATURES)

print("="*60)
print("TRAINING ON REALISTIC DATA")
print("="*60)

# Load data
print("\nLoading dataset...")
df = pd.read_csv(CSV_PATH)

def build_sequences(df):
    X, y = [], []

    for _, group in df.groupby("session_id"):
        group = group.sort_values("timestep")

        if len(group) == TIMESTEPS:
            X.append(group[FEATURES].values)
            y.append(group["label"].iloc[0])

    return np.array(X), np.array(y)

X, y = build_sequences(df)

print(f"\nData shape: X={X.shape}, y={y.shape}")
print(f"Class distribution: {np.bincount(y)}")
print(f"Class percentages: {(np.bincount(y) / len(y) * 100).round(1)}")
```

```python

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.15, stratify=y_train, random_state=42
)

print(f"\nSplit: Train={len(X_train)}, Val={len(X_val)}, Test={len(X_test)}")

# Use RobustScaler instead of StandardScaler (more robust to outliers)
scaler = RobustScaler()

X_train_2d = X_train.reshape(-1, NUM_FEATURES)
X_val_2d = X_val.reshape(-1, NUM_FEATURES)
X_test_2d = X_test.reshape(-1, NUM_FEATURES)

X_train = scaler.fit_transform(X_train_2d).reshape(X_train.shape)
X_val = scaler.transform(X_val_2d).reshape(X_val.shape)
X_test = scaler.transform(X_test_2d).reshape(X_test.shape)
```

```python

# Build model optimized for realistic, overlapping data
def build_model():
    model = Sequential([
        # Input noise to improve generalization
        GaussianNoise(0.05, input_shape=(TIMESTEPS, NUM_FEATURES)),

        # First Bi-LSTM layer
        Bidirectional(LSTM(96, return_sequences=True,
                          kernel_regularizer=l1_l2(l1=0.0005, l2=0.001),
                          recurrent_dropout=0.2)),
        BatchNormalization(),
        Dropout(0.4),

        # Second LSTM layer
        LSTM(64, kernel_regularizer=l1_l2(l1=0.0005, l2=0.001),
             recurrent_dropout=0.2),
        BatchNormalization(),
        Dropout(0.4),

        # Dense layers
        Dense(96, activation="relu",
              kernel_regularizer=l1_l2(l1=0.0005, l2=0.001)),
        BatchNormalization(),
        Dropout(0.35),

        Dense(48, activation="relu",
              kernel_regularizer=l1_l2(l1=0.0003, l2=0.0008)),
        Dropout(0.3),

        # Output
        Dense(3, activation="softmax")
    ])

    # Use slightly lower learning rate for better convergence
    model.compile(
        optimizer=Adam(learning_rate=0.0008),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model

model = build_model()
model.summary()
```

```python
import numpy as np
import joblib
from tensorflow.keras.models import load_model

print("="*60)
print("GAME SAFETY CLASSIFIER - INFERENCE")
print("="*60)

# Load model & scaler
print("\nLoading model...")
model = load_model("lstm_permission_model.keras")
scaler = joblib.load("scaler.pkl")

TIMESTEPS = 30
NUM_FEATURES = 11
LABELS = ["Safe", "Suspicious", "Dangerous"]
```

```python

# ============================================================================
# PREPROCESSING FUNCTIONS
# ============================================================================

def smart_pad_session(session):
    """Pad or truncate session to exactly TIMESTEPS"""
    session = np.array(session, dtype=float)

    if session.ndim != 2 or session.shape[1] != NUM_FEATURES:
        raise ValueError(f"Each timestep must have exactly {NUM_FEATURES} features")

    # Truncate if too long
    if session.shape[0] > TIMESTEPS:
        session = session[:TIMESTEPS]

    # Pad if too short (repeat last timestep)
    if session.shape[0] < TIMESTEPS:
        pad = np.repeat(session[-1][np.newaxis, :],
                        TIMESTEPS - session.shape[0],
                        axis=0)
        session = np.vstack([session, pad])

    return session

def preprocess_session(session):
    """Prepare session for model input"""
    session = smart_pad_session(session)

    # Reshape and scale
    x = session.reshape(1, TIMESTEPS, NUM_FEATURES)
    x_2d = x.reshape(-1, NUM_FEATURES)
    x_scaled = scaler.transform(x_2d).reshape(1, TIMESTEPS, NUM_FEATURES)

    return x_scaled
```

```python

# ============================================================================
# PREDICTION FUNCTION
# ============================================================================

def predict_session(session, verbose=True):
    """
    Predict session safety level

    Args:
        session: List of timesteps, each with 11 features:
                 [cam_count, cam_duration, cam_bg_ratio, loc_freq, loc_bg_ratio,
                  mic_duration, mic_bg_flag, data_upload, data_download, bg_data, reserved]
        verbose: Print detailed results

    Returns:
        Dictionary with prediction and probabilities
    """
    x = preprocess_session(session)

    # Get probabilities
    probs = model.predict(x, verbose=0)[0]

    # Get prediction
    pred_idx = np.argmax(probs)
    pred_label = LABELS[pred_idx]
    confidence = float(probs[pred_idx])

    result = {
        "prediction": pred_label,
        "confidence": round(confidence, 3),
        "probabilities": {
            "Safe": round(float(probs[0]), 3),
            "Suspicious": round(float(probs[1]), 3),
            "Dangerous": round(float(probs[2]), 3)
        }
    }

    if verbose:
        print(f"\nPrediction: {pred_label}")
        print(f"Confidence: {confidence:.1%}")
        print(f"Probabilities:")
        for label, prob in result['probabilities'].items():
            bar = "█" * int(prob * 50)
            print(f"  {label:12s} {prob:.3f} {bar}")

    return result
```

```python

def create_session(cam_count, cam_duration, cam_bg_ratio,
                   loc_freq, loc_bg_ratio,
                   mic_duration, mic_bg_flag,
                   data_upload, data_download, bg_data,
                   timesteps=30):
    """
    Create a session with given feature values

    Features:
        cam_count: Camera access count per timestep
        cam_duration: Camera usage duration (seconds)
        cam_bg_ratio: Ratio of background camera use (0-1)
        loc_freq: Location access frequency
        loc_bg_ratio: Ratio of background location use (0-1)
        mic_duration: Microphone usage duration (seconds)
        mic_bg_flag: Microphone in background (0 or 1)
        data_upload: Data upload amount (KB)
        data_download: Data download amount (KB)
        bg_data: Background data usage (KB)
    """
    session = []
    for _ in range(timesteps):
        session.append([
            cam_count, cam_duration, cam_bg_ratio,
            loc_freq, loc_bg_ratio,
            mic_duration, mic_bg_flag,
            data_upload, data_download, bg_data,
            0  # reserved
        ])
    return session
```
